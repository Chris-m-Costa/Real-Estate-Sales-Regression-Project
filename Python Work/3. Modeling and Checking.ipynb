```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error 
from sklearn.neighbors import KNeighborsRegressor
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
#Supressing a warning about deprecated sns.distplot further down
```

```python
# First we will import and model the baseline data. taken without any triming of outliers or transformations
```

```python
baseline_data = pd.read_csv('baseline_model_data.csv')
baseline_data.head()
```

```python
baseline_data.shape
```

```python
baseline_target = baseline_data['price']
baseline_features = baseline_data.drop('price', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(baseline_features, baseline_target, test_size=.2, random_state=100)
```

```python

LR_model = LinearRegression()
LR_model.fit(X_train, y_train)
```

```python
LR_predictions = LR_model.predict(X_test)
```

```python
r_sq = r2_score(LR_predictions, y_test)
r_sq

#Accuracy score for the baseline data regression model. r_sq scores range from 0-1 for a positive model fit.
# A score closer to 1 is a more accurate model
```

```python
knn_model = KNeighborsRegressor(n_neighbors=3)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_predictions = knn_model.predict(X_test)
```

```python
r_sq = r2_score(knn_predictions, y_test)
r_sq

# The KNN accuracy score is much lower than that of the regression model indicating a poor model fitment to our data.
# This is not an effective model for our applicatuion.
```

```python
# Now the baseline data has been modeled we will run the moels again on our final dataset with outliers trimmed. 
# No other transformations were deemed necessary
```

```python
final_data = pd.read_csv('final_model_data.csv')
final_data.head()
```

```python
final_data.shape
```

```python
final_target = final_data['price']
final_features = final_data.drop('price', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_features, final_target, test_size=.2, random_state=100)
```

```python

LR_model = LinearRegression()
LR_model.fit(X_train, y_train)
```

```python
LR_predictions = LR_model.predict(X_test)
```

```python
r_sq = r2_score(LR_predictions, y_test)
r_sq

# We can see here a modest improvement in the fitment of our regression model over the un-trimmed baseline data
```

```python
knn_model = KNeighborsRegressor(n_neighbors=3)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_predictions = knn_model.predict(X_test)
```

```python
r_sq = r2_score(knn_predictions, y_test)
r_sq

# The KNN model again proving a poor fit, more so this time on the trimmed data. 
```
